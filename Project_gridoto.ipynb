# Groupby

```python
data_gridoto['Spesifikasi'].unique()
```

```python
data_groupby = data_gridoto[['Spesifikasi','Tahun','Harga']]

harga_rata_rata = data_groupby.groupby(['Spesifikasi','Tahun'],as_index=False).mean()
```

```python
harga_rata_rata.head(10)
```

```python
pivot_harga = harga_rata_rata.pivot(index='Spesifikasi', columns='Tahun')

pivot_harga
```

```python
pivot_harga.replace('NaN', '0',inplace=True)

```

```python
pivot_harga.fillna('0',inplace=True)

pivot_harga
```

# Heatmap

```python
pivot_harga = pivot_harga[pivot_harga.columns].astype(float)

sns.heatmap(pivot_harga)

plt.show()
```
